# **Bar Chart**

```python
# Count gender
gender_count = df['gender'].value_counts()

# Simple bar chart
plt.bar(['Female', 'Male'], gender_count.values)
plt.title('Gender Count')
plt.show()
```

```python
# Count abonoment_type
abonoment_type_count = df['abonoment_type'].value_counts()

# Simple bar chart
plt.bar(['Standard', 'Premium'], abonoment_type_count.values)
plt.title('abonoment_type Count')
plt.show()
```

# **Histogram**

```python
# avg_time_in_gym distribution
plt.hist(df['avg_time_in_gym'])
plt.title('avg_time_in_gym')
plt.show()
```

```python
# With more bins
plt.hist(df['avg_time_in_gym'], bins=50, color='green')
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('avg_time_in_gym')
plt.show()
```

# **Line Graph**

```python
# Simple line plot: First 50 count
plt.plot(df['avg_time_in_gym'].head(50))

plt.title('avg_time_in_gym')  # Title
plt.xlabel('Student Number')  # X label
plt.ylabel('avg_time_in_gym')  # Y label
plt.grid(True, which="both", linestyle="-", linewidth=0.7, alpha=0.7)
plt.savefig("/content/drive/MyDrive/Datasets/firstcheck.png",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Datasets/firstcheck.jpg",dpi=300, bbox_inches='tight')
plt.savefig("/content/drive/MyDrive/Datasets/firstcheck.pdf",dpi=300, bbox_inches='tight')
plt.show()  # Show plot
```

# **Scatter Plot**

```python
# Age vs avg_time_in_gym
plt.scatter(df['Age'], df['avg_time_in_gym'])
plt.grid(True)
plt.xlabel('Age')
plt.ylabel('avg_time_in_gym')
plt.title('Age vs avg_time_in_gym')
plt.show()
```
